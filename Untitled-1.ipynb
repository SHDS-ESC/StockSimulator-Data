```python
import mysql.connector

# DB 연결 정보 설정
# host: 로컬 PC이므로 'localhost' 또는 '127.0.0.1'
# user: MariaDB 사용자 이름
# password: 사용자 비밀번호
# database: 접속할 데이터베이스 이름
config = {
    'host': '127.0.0.1',
    'user': 'testuser',
    'password': 'test1234',
    'database': 'stock_simul'
}

# 변수 선언
db_connection = None
cursor = None

try:
    # 데이터베이스 연결
    db_connection = mysql.connector.connect(**config)
    
    # 연결 확인
    if db_connection.is_connected():
        print("Successfully connected to MariaDB.")

    # 커서 생성 (SQL 쿼리를 실행하는 객체)
    cursor = db_connection.cursor()
    
    # 예제 SQL 쿼리 실행 (예: 'users' 테이블에서 모든 데이터 가져오기)
    query = "SELECT * FROM stock;"
    cursor.execute(query)

    # 모든 결과 가져오기
    results = cursor.fetchall()
    
    # 결과 출력
    if results:
        for row in results:
            print(row)
    else:
        print("No data found.")

except mysql.connector.Error as e:
    print(f"Error connecting to MariaDB: {e}")

finally:
    # 연결 및 커서 닫기 (매우 중요)
    if cursor is not None:
        cursor.close()
    if db_connection is not None:
        db_connection.close()
        print("Connection closed.")
```

```python
# 예시 코드
import pandas as pd

# 'actions' 데이터프레임
actions_df = pd.DataFrame({
    'date': pd.to_datetime(['1976-03-31', '2023-08-28', '2006-03-31', '1982-11-30']),
    'ticker': ['IFF', 'WBA', 'GOOG', 'AAPL'],
    'action': ['added', 'added', 'added', 'added']
}).sort_values('date').reset_index(drop=True)

# 초기 S&P 500 목록 (예시)
sp500_list = {'A', 'B', 'C', 'D'}

# 날짜별 종목 상태를 저장할 딕셔너리
sp500_history = {}

# 날짜 순으로 상태 업데이트
for index, row in actions_df.iterrows():
    current_date = row['date']
    ticker = row['ticker']
    action = row['action']

    # 이전 날짜의 목록을 복사하여 현재 날짜의 목록으로 초기화
    # 딕셔너리에 데이터가 없으면 초기 목록을 사용
    if not sp500_history:
        sp500_history[current_date.date()] = sp500_list.copy()
    else:
        last_date = max(sp500_history.keys())
        sp500_history[current_date.date()] = sp500_history[last_date].copy()

    # 'action'에 따라 종목 추가/제거
    if action == 'added':
        sp500_history[current_date.date()].add(ticker)
    elif action == 'removed':
        sp500_history[current_date.date()].remove(ticker)

# 결과 출력 (예시)
print(sp500_history)
```
