```python
import yfinance as yf
import json
import pprint

# 티커 객체 생성
ticker_symbol = "PXD"
stock_info = yf.Ticker(ticker_symbol)

# .info 속성으로 기업 정보 가져오기
company_info = stock_info.info

# 파일로 저장하기 (JSON 형식으로 저장하면 나중에 읽기 편합니다)
file_path = f"{ticker_symbol}_info.json"

with open(file_path, "w", encoding="utf-8") as f:
    # pprint.pformat()을 사용하면 문자열 형태로 포맷팅하여 파일에 쓸 수 있습니다.
    pprint_string = pprint.pformat(company_info, indent=4)
    f.write(pprint_string)

print(f"모든 정보가 '{file_path}' 파일에 저장되었습니다.")

# 또는 JSON 모듈을 직접 사용해도 됩니다.
# with open(f"{ticker_symbol}_info_json.json", "w", encoding="utf-8") as f:
#     json.dump(company_info, f, indent=4, ensure_ascii=False)
```
